### Inizializzare un grafico

### Personalizzare un grafico

```python
x = np.arange(25,31)
y = [100, 120, 140, 110, 150, 145]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(25,31)
y_1 = [100, 120, 140, 110, 150, 145]
y_2 = [90, 80, 120, 150, 170, 140]
plt.plot(x,y_1, label='Italiani', color='red', linewidth=2, marker='o')
plt.plot(x,y_2, label='Francesi', color='green', linestyle='dotted')
plt.title('Nome del grafico')
plt.xlabel('Eta')
plt.ylabel('Valori')
plt.legend()
plt.grid()
plt.show()
```

### Grafico a barre

```python
x = np.arange(25,31)
y_1 = [100, 120, 140, 110, 150, 145]
y_2 = [90, 80, 120, 150, 170, 140]
plt.bar(x,y_1, label='Italiani', color='red')
plt.plot(x,y_2, label='Francesi', color='black', linestyle='dotted', linewidth=3)
plt.bar(x,y_2, label='Francesi', color='black')
# posso mettere assieme piu' tipi di grafici diversi

plt.title('Nome del grafico')
plt.xlabel('Eta')
plt.ylabel('Valori')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.arange(25,31)
larghezza = 0.3
y_1 = [100, 120, 140, 110, 150, 145]
y_2 = [90, 80, 120, 150, 170, 140]
plt.bar(x+larghezza/2,y_1, width=larghezza, label='Italiani', color='red')
plt.bar(x-larghezza/2,y_2, width=larghezza, label='Francesi', color='blue')

plt.title('Nome del grafico')
plt.xlabel('Eta')
plt.ylabel('Valori')
plt.legend()
plt.grid()
plt.show()
```

### Grafico a torta

```python
slices = [200,134,26]
etichette = ['si', 'no', 'non lo so']
colori = ['green', 'red', 'yellow']   # i colori possono essere aggiunti in esadecimale
esplosione = [0,0,0.5]
plt.pie(slices, labels=etichette, colors=colori, wedgeprops={'edgecolor': 'black'}, startangle=45, explode=esplosione, autopct='%1.1f%%')
plt.title('Grafico a torta')
plt.show()
```

### Lavorare con le date

```python
x = [dt(2022,1,10), dt(2022,1,15), dt(2022,1,20), dt(2022,1,25), dt(2022,1,30)]
y = [3,6,2,8,5]
plt.plot_date(x,y, linestyle='solid')
plt.gcf().autofmt_xdate()
date_formattate = mpl_dates.DateFormatter('%d %b %Y')
plt.gca().xaxis.set_major_formatter(date_formattate)
plt.grid()
plt.show()
```

### Subplot

```python
plt.style.use('seaborn')




data = pd.read_csv('data_salary_CoreySchafer.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']




fig, (ax1, ax2) = plt.subplots(1,2)   # create a figure that can contain how many axes we want

ax1.plot(ages, py_salaries, label='Python')
ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax1.legend()
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')
ax1.set_title('Python')


ax2.plot(ages, js_salaries, label='JavaScript')
ax2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')
ax2.set_title('JavaScript')

fig.suptitle('Median Salary (USD) by Age', fontsize=20)
fig.tight_layout()




fig2, (ax1_2, ax2_2) = plt.subplots(2,1, sharex=True)   # create a figure that can contain how many axes we want

ax1_2.plot(ages, py_salaries, label='Python')
ax1_2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax1_2.legend()
ax1_2.set_ylabel('Median Salary (USD)')
ax1_2.set_title('Python')


ax2_2.plot(ages, js_salaries, label='JavaScript')
ax2_2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
ax2_2.legend()
ax2_2.set_xlabel('Ages')
ax2_2.set_ylabel('Median Salary (USD)')
ax2_2.set_title('JavaScript')

fig2.suptitle('Median Salary (USD) by Age', fontsize=20)
fig2.tight_layout()




plt.show()


```
