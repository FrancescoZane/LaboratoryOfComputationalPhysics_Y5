Chiaramente prima è necessario studiare l'effetto del campo magneto sui muoni, i quali entrano con un momento definito in un punto definito nello spazio. (consideriamo il punto x,y nel piano trasversale all'esperimento, mentre Z è fissato dalla posizione del magnete).

```python
B=1.7 #Tesla
L=3 #meters  

#########################
#### Positive muons #####
#########################

#initial momenta and circ radiuses
P_pos=np.sqrt(Px_mu2**2+Pz_mu2**2)
R_pos=P_pos/(B*0.3) #meters
Circ_center=np.empty((NEvents,2))

#initial positions
X_init_pos=
Y_init_pos=
Z_init_pos=

#Circ center
alpha=np.pi/2 - np.arccos(Pz_mu2/P_pos)

Circ_center[:,0]=X_init_pos+R_pos*np.cos(alpha)
Circ_center[:,1]=Z_init_pos+R_pos*np.sin(alpha)

#positions after the magnet
Pos_mupos=np.empty((NEvents,3))

Pos_mupos[:,0]= Circ_center[:,0]+np.sqrt(R_pos**2-((Z_init_pos+L)-Circ_center[:,1])**2)
Pos_mupos[:,2]= Z_init_pos+L

corda=np.sqrt((Pos_mupos[:,0]-X_init_pos)**2+(Pos_mupos[:,2]-Z_init_pos)**2)
theta=2*np.arcsin(corda/(2*R_pos))

k=R_pos*theta/P_pos

Pos_mupos[:,1]=Py_mu2*k+Y_init_pos   

#momenta
Momenta_mupos=np.empty((NEvents,2))
Momenta_mupos[:,0]=np.cos(theta)*Px_mu2-np.sin(theta)*Pz_mu2
Momenta_mupos[:,1]=np.cos(theta)*Pz_mu2-np.sin(theta)*Px_mu2

Px_mu2=Momenta_mupos[:,0]
Py_mu2=Py_mu2
Pz_mu2=Momenta_mupos[:,1]

#########################
#### Negative muons #####
#########################

#initial momenta and circ radiuses
P_neg=np.sqrt(Px_mu1**2+Pz_mu1**2)
R_neg=P_neg/(B*0.3) #meters
Circ_center=np.empty((NEvents,2))

#initial positions
X_init_neg=
Y_init_neg=
Z_init_neg=

#Circ center
alpha=np.pi/2 - np.arccos(Pz_mu1/P_neg)

Circ_center[:,0]=X_init_neg-R_neg*np.cos(alpha)
Circ_center[:,1]=Z_init_neg+R_neg*np.sin(alpha)

#positions after the magnet
Pos_muneg=np.empty((NEvents,3))

Pos_muneg[:,0]= Circ_center[:,0]-np.sqrt(R_neg**2-((Z_init_neg+L)-Circ_center[:,1])**2)
Pos_muneg[:,2]= Z_init_neg+L

corda=np.sqrt((Pos_muneg[:,0]-X_init_neg)**2+(Pos_muneg[:,2]-Z_init_neg)**2)
theta=2*np.arcsin(corda/(2*R_neg))

k=R_neg*theta/P_neg

Pos_muneg[:,1]=Py_mu1*k+Y_init_neg  

#momenta
Momenta_muneg=np.empty((NEvents,2))
Momenta_muneg[:,0]=np.cos(-theta)*Px_mu1-np.sin(-theta)*Pz_mu1
Momenta_muneg[:,1]=np.cos(-theta)*Pz_mu1-np.sin(-theta)*Px_mu1

Px_mu1=Momenta_muneg[:,0]
Py_mu1=Py_mu1
Pz_mu1=Momenta_muneg[:,1]
```

Definizione della funzione che si occupa del tracciare l'esperimento. Questa restituisce solo gli hits.

```python
def hits(total_datf,numberf):
    
    #splitting input matrix for convenience
    (Xf, Yf, Zf, Px_ef, Py_ef, Pz_ef, Px_mu1f, Py_mu1f, Pz_mu1f, Px_mu2f, Py_mu2f, Pz_mu2f,
            Pos_muneg0f, Pos_muneg1f, Pos_muneg2f, Momenta_muneg0f, Momenta_muneg1f,
            Pos_mupos0f, Pos_mupos1f, Pos_mupos2f, Momenta_mupos0f, Momenta_mupos1f) = (total_datf)
    
    #number of external signals due to noise
    numberf=numberf
    
    #posizione del detector
    Z_det = np.array([-0.21,-0.11,-0.01,
                      5.03,6.03,7.03,
                      9.43,10.43,11.43])
    
    #punto di riferimento del detector
    det_XY = np.array([[-0.011,-0.011], [-0.011,-0.011], [-0.011,-0.011],
                       [-0.025,-0.025], [-0.028,-0.028], [-0.03,-0.03],
                       [0.009,-0.037],  [0.039,-0.04],   [0.068,-0.043]])
    
    #dimensioni laterali del detector
    det_sides = np.array([[0.022, 0.022], [0.022, 0.022], [0.022, 0.022],
                          [0.05, 0.05], [0.056, 0.056], [0.06, 0.06],
                          [0.094, 0.074], [0.128, 0.08],[0.164, 0.086]])
    
    #lists used for saving coordinates of active pixels' centers (i.e. both noise and signal)
    #in every list there will be three elements, one for each detector of each bunch
    #(before target, between target and magnetic field, after magnetic field "right" and "left")
    #every element will be a list of coordinates, so 2-elements lists
    hit_raw_before = []
    hit_raw_between = []
    hit_raw_aftern = []
    hit_raw_afterp = []
    
    """
    Simulation before the target - positrons
    """
    
    for k in range(0,3):
        #EOM
        i=Z_det[k]
        t=(i-Zf)/Pz_ef   
        x_sim=Xf+Px_ef*t
        y_sim=Yf+Py_ef*t
        
        #assignment of new names for convenience
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[k][0],det_XY[k][1],det_sides[k][0],det_sides[k][1],50e-6,x_sim,y_sim)
        
        
        #number is a matrix with poisson casual numbers that represent the number n of active pixels
        #in a single detector due to noise
        #so we extract n couple of integer numbers that are the coordinates of the noisy pixels
        #expressed in pixels-unit
        
        temp1=np.random.randint(1,np.int(sideX/pitch),numberf[k])
        temp2=np.random.randint(1,np.int(sideY/pitch),numberf[k])
        
        #Ora vogliamo esprimere le coordinate dei pixel rumorosi in metri
        pixels = [[0.1, 0.1] for i in range(len(temp1))] #dummy pixels list of the right length
        
        for i in range(0,len(temp1)):
            #saving coordinates of noisy pixels
            pixels[i] = [X0+temp1[i]*pitch-0.5*pitch,Y0+temp2[i]*pitch-0.5*pitch]
        
        #Ora viene l'esperimento vero e proprio. Qui facciamo passare N positroni e verifichiamo se passano
        #nel detector o meno.
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            #checking for particle being within detector
            pixels.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        #elenco pixel eccitati per ogni detector (ogni riga dell'array)
        hit_raw_before.append(pixels)
        
        
    """
    Simulation between the target and the magnet
    """
    
    for k in range(3,6):
        i=Z_det[k]
        tn=(i-Zf)/Pz_mu1f   
        xn=Xf+Px_mu1f*tn    
        yn=Yf+Py_mu1f*tn
        tp=(i-Zf)/Pz_mu2f
        xp=Xf+Px_mu2f*tp
        yp=Yf+Py_mu2f*tp
        
        ### negative muons ###
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[k][0],det_XY[k][1],det_sides[k][0],det_sides[k][1],50e-6,xn,yn)
        temp1=np.random.randint(1,np.int(sideY/pitch),numberf[k])
        temp2=np.random.randint(1,np.int(sideX/pitch),numberf[k])
        pixels1 = [[0.1, 0.1] for i in range(len(temp1))]
        
        for i in range(0,len(temp1)):
            pixels1[np.intp(i)] = [X0+temp2[np.intp(i)]*pitch-0.5*pitch,Y0+temp1[np.intp(i)]*pitch-0.5*pitch]
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels1.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        ### positive muons ###
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[k][0],det_XY[k][1],det_sides[k][0],det_sides[k][1],50e-6,xp,yp)
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels1.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        #elenco pixel eccitati per ogni detector (ogni riga dell'array)
        hit_raw_between.append(pixels1)
          
      
    """
    Simulation between the target and the magnet
    """
    
    #mu-
    for k in range(6,9):
        i=Z_det[k]
        tn=(i-Pos_muneg2f)/Momenta_muneg1f
        xn=Pos_muneg0f+Momenta_muneg0f*tn
        yn=Pos_muneg1f+Py_mu1f*tn
        
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[np.intp(k)][0],det_XY[np.intp(k)][1],det_sides[np.intp(k)][0],det_sides[np.intp(k)][1],50e-6,xn,yn)
        temp1=np.random.randint(1,np.int(sideX/pitch),numberf[np.intp(k)])
        temp2=np.random.randint(1,np.int(sideY/pitch),numberf[np.intp(k)])
        pixels2 = [[0.1, 0.1] for i in range(len(temp1))]
        
        for i in range(0,len(temp1)):
            pixels2[np.intp(i)] = [X0+temp1[np.intp(i)]*pitch-0.5*pitch,Y0+temp2[np.intp(i)]*pitch-0.5*pitch]
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels2.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        hit_raw_aftern.append(pixels2)
    
    #mu+
    for k in range(6,9):
        i=Z_det[k]
        tp=(i-Pos_mupos2f)/Momenta_mupos1f
        xp=Pos_mupos0f+Momenta_mupos0f*tp
        yp=Pos_mupos1f+Py_mu2f*tp
        
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(-det_XY[np.intp(k)][0]-det_sides[np.intp(k)][0],det_XY[np.intp(k)][1],det_sides[np.intp(k)][0],det_sides[np.intp(k)][1],50e-6,xp,yp)
        temp1=np.random.randint(1,np.int(sideY/pitch),numberf[np.intp(k)])
        temp2=np.random.randint(1,np.int(sideX/pitch),numberf[np.intp(k)])
        pixels3 = [[0.1, 0.1] for i in range(len(temp1))]
        
        for i in range(0,len(temp1)):
            pixels3[np.intp(i)] = [X0+temp2[np.intp(i)]*pitch-0.5*pitch,Y0+temp1[np.intp(i)]*pitch-0.5*pitch]
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels3.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        hit_raw_afterp.append(pixels3)
        
    return hit_raw_before, hit_raw_between, hit_raw_aftern, hit_raw_afterp
```
