## Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import poisson
from skspatial.objects import Line, Points
from skspatial.plotting import plot_3d
import scipy.odr as spo


#from skspatial.objects import Line, Points
#from skspatial.plotting import plot_3d

df=pd.read_csv("dataset.txt", sep=" ", index_col=False)

#Removing unphysical events with just zeros by asking the momenta of the positron to be bigger than 1 GeV
df=df[df.iloc[:,11]>1] 

df
```

```python
mu_mass = 105.65/(10**3)
c = 299792458

NEvents=df.shape[0]

X=df["x_scattering"]
Y=df["y_scattering"]
Z=df["z_scattering"]

px_e=df["px_positron"]
py_e=df["py_positron"]
pz_e=df["pz_positron"]

Px_mu1=df["px_mu1"]
Py_mu1=df["py_mu1"]
Pz_mu1=df["pz_mu1"]

Px_mu2=df["px_mu2"]
Py_mu2=df["py_mu2"]
Pz_mu2=df["pz_mu2"]
```

## Experiment dimensions

### Before the target - positrons

```python
#Z pos of the detectors before the target
Pos1= -0.21
Pos2= -0.11
Pos3= -0.01

Z_Det_beforeTarg=[Pos1,Pos2,Pos3]
Position_e_p = pd.DataFrame()

for i in Z_Det_beforeTarg:
    t=(i-Z)/(pz_e)   
    x=X+(px_e)*t  
    y=Y+(py_e/mu_mass*c)*t
    
    d = np.sqrt(x**2+y**2)     #distance from the z-axis
    d_M =np.max(d)             #let's take the max distance 
    
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(18,8))
    fig.suptitle('Position distribution at Z={} meters before target'.format(-1*i),fontsize='x-large')
    
    #plot the histogram with the particle distribution
    h2d = ax1.hist2d(x,y,bins=(100,100))
    ax1.set_xlabel("Position along x (m)",fontsize='large')
    ax1.set_ylabel("Position along y (m)",fontsize='large')
    fig.colorbar(h2d[3], ax=ax1)
    
    #distribution with the detector shape
    ax2.scatter(x,y,color='red', marker='+')
    
    #detector figure
    r= d_M*1.1         #radius
    ax2.vlines(-r, -r, r)
    ax2.vlines(r, -r, r)
    ax2.hlines(-r, -r, r)
    ax2.hlines(r, -r, r)
    
    ax2.set_xlabel("Position along x (m)",fontsize='large')
    ax2.set_ylabel("Position along y (m)",fontsize='large')
    plt.show()
    
    print("In order to contain the beam, we can use a detector with a side of {:.3f} m".format(2*r))
    Position_e_p['e+_X_{}'.format(-1*i)] = x
    Position_e_p['e+_Y_{}'.format(-1*i)] = y
    Position_e_p['e+_Z_{}'.format(-1*i)] = i*np.ones(NEvents)
```

As we can see, the positron beam is not devergent since px and py are vanishing.

### Between the target and the magnet - muon pairs

Now, we have to treat the experiment between the target and the magnet, where muons and antimuons are not well separated in space because they are produced in the same point of space-time. Since we want to study them separately, we have to take into account that we need pixels with a size capable of separating these particles (study of smearing is needed).

```python
#detector positions
Z_Det_beforeMagn=[5.03, 6.03, 7.03]
Position_mu_n = pd.DataFrame()
Position_mu_p = pd.DataFrame()

for i in Z_Det_beforeMagn:
    
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(18,8))
    fig.suptitle('Muons and antimuons positions along the trasversal plane at Z={} meters from target'.format(i),fontsize='x-large')
    
    ######################
    ### negative muons ###
    ######################
    
    tn=(i-Z)/(Pz_mu1/mu_mass*c)   
    xn=X+(Px_mu1/mu_mass*c)*tn    
    yn=Y+(Py_mu1/mu_mass*c)*tn
       
    #particle positions
    ax1.scatter(xn,yn,color='blue',marker='+')
    
    #detector boundaries
    d = np.sqrt(xn**2+yn**2)
    r = 1.1 * np.max(d)
    
    ax1.vlines(-r, -r, r)
    ax1.vlines(r, -r, r)
    ax1.hlines(-r, -r, r)
    ax1.hlines(r, -r, r)
    
    
    ax1.set_xlabel("Muons Position along x (m)")
    ax1.set_ylabel("Muons Position along y (m)")
    
    Position_mu_n['mu-_X_{}'.format(i)] = xn
    Position_mu_n['mu-_Y_{}'.format(i)] = yn
    Position_mu_n['mu-_Z_{}'.format(i)] = i*np.ones(NEvents)
    
    ######################
    ### positive muons ###
    ######################
    
    tp=(i-Z)/Pz_mu2
    xp=X+Px_mu2*tp
    yp=Y+Py_mu2*tp
    
    #particle positions
    ax2.scatter(xp,yp,color='red',marker='+')
    
    #detector boundaries
    d = np.sqrt(xp**2+yp**2)
    r = 1.1 * np.max(d)
    
    ax2.vlines(-r, -r, r)
    ax2.vlines(r, -r, r)
    ax2.hlines(-r, -r, r)
    ax2.hlines(r, -r, r)
    
    
    ax2.set_xlabel("Antimuons Position along x (m)")
    ax2.set_ylabel("Antimuons Position along y (m)")
    
    Position_mu_p['mu+_X_{}'.format(i)] = xp
    Position_mu_p['mu+_Y_{}'.format(i)] = yp
    Position_mu_p['mu+_Z_{}'.format(i)] = i*np.ones(NEvents)
    
    plt.show()
    
    print("In order to contain the beam, we can use a detector with a side of {:.3f} m".format(2*r))
    
    #########################
    ### study of smearing ###
    #########################

    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(18,8))
    fig.suptitle('Distance between muons of same event at Z={} meters from target'.format(i),fontsize='x-large')
    
    dist=np.sqrt((xp-xn)**2+(yp-yn)**2)
    
    ax1.hist(dist,bins=100,alpha = 0.5, lw=3)
    ax1.set_xlabel("Distance (m)")
    ax1.set_ylabel("Counts")

    #histogram stopped at 50 Âµm
    ax2.hist(dist,range=[0,0.00005],alpha = 0.5, lw=3)
    ax2.set_xlabel("Distance (m)")
    ax2.set_ylabel("Counts")
    plt.show()
```

It is shown that we can use detectors with the following geometrical stats:
 1. Z=5.03 m - side=0.041 m
 2. Z=6.03 m - side=0.044 m
 3. Z=7.03 m - side=0.048 m

Moreover, taking 50 Âµm as the pixel size, it is verified that it is possible to separate approximately 100% of the pairs. We can reduce the number of unseparated pair placing the first detector far from the target, but this is a solution that doen't take into account that the farther away we put the detectors, the larger they have to be to have 100% acceptance.

### After the magnet - muon beams

First of all we have to evaluate the cinematic properties of the muons and the antimuons after the magnet

```python
B=1.7 #Tesla
L=3 #meters  
Z_mag = 8.03

#########################
#### Positive muons #####
#########################

#initial momenta and circ radiuses
P_pos=np.sqrt(Px_mu2**2+Pz_mu2**2)
R_pos=P_pos/(B*0.3)          
Circ_center=np.zeros((NEvents,2))

#initial positions
Z_init_pos=np.ones(NEvents)*Z_mag
t=(Z_init_pos-Z)/Pz_mu2

X_init_pos=X+Px_mu2*t
Y_init_pos=Y+Py_mu2*t

#Circ center
alpha=np.arccos(Pz_mu2/P_pos)
beta=np.pi/2-alpha

Circ_center[:,0][Px_mu2>=0]=(X_init_pos+R_pos*np.sin(beta))[Px_mu2>=0]
Circ_center[:,1][Px_mu2>=0]=(Z_init_pos-R_pos*np.cos(beta))[Px_mu2>=0]
   
Circ_center[:,0][Px_mu2<0]=(X_init_pos+R_pos*np.sin(beta))[Px_mu2<0]
Circ_center[:,1][Px_mu2<0]=(Z_init_pos+R_pos*np.cos(beta))[Px_mu2<0]

#positions after the magnet
Pos_mupos=np.zeros((NEvents,3))

Pos_mupos[:,0]= Circ_center[:,0]-np.sqrt(R_pos**2-((Z_init_pos+L)-Circ_center[:,1])**2)  # x
Pos_mupos[:,2]= Z_init_pos+L    # z

string=np.sqrt((Pos_mupos[:,0]-X_init_pos)**2+(Pos_mupos[:,2]-Z_init_pos)**2)
theta=2*np.arcsin(string/(2*R_pos))
theta=2*np.arcsin(3/(2*R_pos))

k=R_pos*theta/P_pos
Pos_mupos[:,1]=Py_mu2*k+Y_init_pos    # y  

#momenta
Momenta_mupos=np.zeros((NEvents,3))

Momenta_mupos[:,0]=np.cos(theta/2)*Px_mu2+np.sin(theta/2)*Pz_mu2
Momenta_mupos[:,1]=Py_mu2
Momenta_mupos[:,2]=np.cos(theta/2)*Pz_mu2-np.sin(theta/2)*Px_mu2



#########################
#### Negative muons #####
#########################

#initial momenta and circ radiuses
P_neg=np.sqrt(Px_mu1**2+Pz_mu1**2)
R_neg=P_neg/(B*0.3)
Circ_center=np.empty((NEvents,2))

#initial positions
Z_init_neg=np.ones(NEvents)*Z_mag
t=(Z_init_neg-Z)/Pz_mu1

X_init_neg=X+Px_mu1*t
Y_init_neg=Y+Py_mu1*t

#Circ center
alpha=np.arccos(Pz_mu1/P_neg)
beta=np.pi/2-alpha

Circ_center[:,0][Px_mu1>=0]=(X_init_neg-R_neg*np.sin(beta))[Px_mu1>=0]
Circ_center[:,1][Px_mu1>=0]=(Z_init_neg+R_neg*np.cos(beta))[Px_mu1>=0]
   
Circ_center[:,0][Px_mu1<0]=(X_init_neg-R_neg*np.sin(beta))[Px_mu1<0]
Circ_center[:,1][Px_mu1<0]=(Z_init_neg-R_neg*np.cos(beta))[Px_mu1<0]

#positions after the magnet
Pos_muneg=np.zeros((NEvents,3))

Pos_muneg[:,0]= Circ_center[:,0]+np.sqrt(R_neg**2-((Z_init_neg+L)-Circ_center[:,1])**2)  # x
Pos_muneg[:,2]= Z_init_neg+L    # z

string=np.sqrt((Pos_muneg[:,0]-X_init_neg)**2+(Pos_muneg[:,2]-Z_init_neg)**2)
theta=2*np.arcsin(string/(2*R_neg))
theta=2*np.arcsin(3/(2*R_neg))

k=R_neg*theta/P_neg
Pos_muneg[:,1]=Py_mu1*k+Y_init_neg    # y  

#momenta
Momenta_muneg=np.zeros((NEvents,3))

Momenta_muneg[:,0]=np.cos(-theta/2)*Px_mu1+np.sin(-theta/2)*Pz_mu1
Momenta_muneg[:,1]=Py_mu1
Momenta_muneg[:,2]=np.cos(-theta/2)*Pz_mu1-np.sin(-theta/2)*Px_mu1
```

Now we can use the EOM of the muons and the antimuons in order to complete the geometrical definition of the experiment.

```python
Position_mu_n_after_mag = pd.DataFrame()
Position_mu_p_after_mag = pd.DataFrame()
```

```python
##solve same equations as before to get where they hit final detectors after magnet
Z_Det_afterMagn=[12.43, 13.43, 14.43] 

s=0
for i in Z_Det_afterMagn:

    #negative muons
    tn=(i-Pos_muneg[:,2])/Momenta_muneg[:,2]   
    
    xn=Pos_muneg[:,0]+Momenta_muneg[:,0]*tn   
    yn=Pos_muneg[:,1]+Momenta_muneg[:,1]*tn
    
    #positive muons
    tp=(i-Pos_mupos[:,2])/Momenta_mupos[:,2] 
    
    xp=Pos_mupos[:,0]+Momenta_mupos[:,0]*tp
    yp=Pos_mupos[:,1]+Momenta_mupos[:,1]*tp
    
    Position_mu_n_after_mag['mu-_X_{}'.format(i)] = xn
    Position_mu_n_after_mag['mu-_Y_{}'.format(i)] = yn
    Position_mu_n_after_mag['mu-_Z_{}'.format(i)] = i*np.ones(NEvents)
    
    Position_mu_p_after_mag['mu+_X_{}'.format(i)] = xn
    Position_mu_p_after_mag['mu+_Y_{}'.format(i)] = yn
    Position_mu_p_after_mag['mu+_Z_{}'.format(i)] = i*np.ones(NEvents)
    
    #detector shape and center
    low_edge=np.array([[0.075,-0.035],[0.10,-0.035],[0.12,-0.035]])
    shape=np.array([[0.15,0.07],[0.175,0.07],[0.2,0.07]])
    
    fig, ax=plt.subplots(1,1,figsize=(18,9))
    fig.suptitle('Muons positions along the trasversal plane at Z={} meters from target'.format(i), fontsize=14)
    
    #antimuon detectors
    ax.vlines(low_edge[s][0], 
              low_edge[s][1], 
              low_edge[s][1]+shape[s][1])
    
    ax.vlines(low_edge[s][0]+shape[s][0],
              low_edge[s][1], 
              low_edge[s][1]+shape[s][1])
    
    ax.hlines(low_edge[s][1],
              low_edge[s][0],
              low_edge[s][0]+shape[s][0])
    ax.hlines(low_edge[s][1]+shape[s][1],
              low_edge[s][0],
              low_edge[s][0]+shape[s][0])
    
    #muon detectors
    ax.vlines(-low_edge[s][0], 
              low_edge[s][1], 
              low_edge[s][1]+shape[s][1])
    
    ax.vlines(-low_edge[s][0]-shape[s][0],
              low_edge[s][1], 
              low_edge[s][1]+shape[s][1])
    
    ax.hlines(low_edge[s][1],
              -low_edge[s][0],
              -low_edge[s][0]-shape[s][0])
    ax.hlines(low_edge[s][1]+shape[s][1],
              -low_edge[s][0],
              -low_edge[s][0]-shape[s][0])
    
    ax.scatter(xn,yn,color='blue',marker='x',label="$\mu^{-}$")
    ax.scatter(xp,yp,color='red',marker='x',label="$\mu^{+}$")
    
    ticksx=np.arange(-(0.25+s*0.05),0.3+s*0.05,0.05)
    ax.set_xticks(ticksx)
    ax.set_xlabel("Position along x (m)")
    ax.set_ylabel("Position along y (m)")
    ax.grid(which='both', linewidth=1)
    ax.legend(loc=2,prop={'size': 20})
    s+=1
    
    plt.show()
```

After this final test, we can see that the geometrical stats of the detectors are the following (meters):

* Muons:
 1. center= ( -0.15 , 0 )     ,     shape=( 0.15 , 0.07 )
 2. center= ( -0.1875 , 0 )   ,     shape=( 0.175 , 0.07 )
 3. center= ( -0.22 , 0 )     ,     shape=( 0.2 , 0.07 )


* Antimuons:
 1. center= ( 0.15 , 0 )      ,     shape=( 0.15 , 0.07 )
 2. center= ( 0.1875 , 0 )    ,     shape=( 0.175 , 0.07 )
 3. center= ( 0.22 , 0 )      ,     shape=( 0.2 , 0.07 )

## Resolution

```python
def hit_detector(data, c, l, p):
    hit = np.array(data)
    det_axis = np.arange(c-l/2, c+l/2+p, p)
    
    matrix_hit = hit.reshape(hit.shape[0],1)*np.ones(det_axis.shape[0])
#     print(matrix_hit.shape)
    matrix_det = np.ones(hit.shape[0]).reshape(hit.shape[0],1) * det_axis
#     print(matrix_det.shape)
    hit_bins = matrix_det[np.abs(matrix_det-matrix_hit)<(p/2)]
    
    return (hit_bins)
```

```python
def tracking(data, z_det, centers_det, size_det, precision_det=0.00005):
    
    k = poisson.rvs(mu=1, size=3)
    ind = [0, k[0], k[0]+k[1], k.sum()]

    noise_xy = np.zeros((k.sum(),2))

    for i in range(3):
        noise_xy[ind[i]:ind[i+1], 0] = np.random.choice(np.arange(-size_det[i]+centers_det[i,0],
                                                                  size_det[i]+centers_det[i,0],
                                                                  precision_det), size=k[i])
        noise_xy[ind[i]:ind[i+1], 1] = np.random.choice(np.arange(-size_det[i]+centers_det[i,1],
                                                                  size_det[i]+centers_det[i,1],
                                                                  precision_det), size=k[i])

    noise_z = np.hstack((z_det[0]*np.ones(k[0]), z_det[1]*np.ones(k[1]), z_det[2]*np.ones(k[2])))

    noise = np.hstack((noise_xy, noise_z.reshape(noise_z.shape[0],1)))

    hit = np.vstack((data.reshape(3, 3), noise))

    det_1 = hit[hit[:,2]==z_det[0]]
    det_2 = hit[hit[:,2]==z_det[1]]
    det_3 = hit[hit[:,2]==z_det[2]]
    
    def f(B, x):
    #         '''Linear function y = m*x + b'''
        return B[0]*x + B[1]
    
    sx = sy = precision_det/2
    sz = 10**(-9)
    
    err = 10**10
    slope = np.ones(2)
    intercept = np.ones(2)
    err_slope = np.ones(2)
    err_int = np.ones(2)
    
    for n1 in range(det_1.shape[0]):
        for n2 in range(det_2.shape[0]):
            for n3 in range(det_3.shape[0]):
                points = np.vstack((det_1[n1, :], det_2[n2, :], det_3[n3, :]))
                x = points[:, 0]
                y = points[:, 1]
                z = points[:, 2]

                linear = spo.Model(f)

                mydata_x = spo.RealData(x, y, sx=sx, sy=sy)
                myodr_x = spo.ODR(mydata_x, linear, beta0=[1., 2.])
                myoutput_x = myodr_x.run()

                mydata_z = spo.RealData(z, y, sx=sz, sy=sy)
                myodr_z = spo.ODR(mydata_z, linear, beta0=[1., 2.])
                myoutput_z = myodr_z.run()
                    
                err_sqrt = np.sqrt(myoutput_z.sum_square + myoutput_x.sum_square)
   
                if err_sqrt<err:
                    err = err_sqrt
                    slope = np.array([myoutput_z.beta[0], myoutput_z.beta[0]/myoutput_x.beta[0]])
                    intercept = np.array([(myoutput_z.beta[1]-myoutput_x.beta[1])/myoutput_x.beta[0],
                                          myoutput_z.beta[1]])
                    err_slope = np.array([myoutput_z.sd_beta[0],
                                         np.sqrt((myoutput_z.sd_beta[0]/myoutput_x.beta[0])**2+
                                                 (myoutput_z.beta[0]*myoutput_x.sd_beta[0]/(myoutput_x.beta[0])**2)**2)])
                    err_int = np.array([myoutput_z.sd_beta[1],
                                       np.sqrt((1/myoutput_x.beta[0])**2*(myoutput_x.sd_beta[1]**2+myoutput_z.sd_beta[1]**2)+
                                               ((myoutput_z.beta[1]-myoutput_x.beta[1])*myoutput_x.sd_beta[0]/(myoutput_x.beta[0])**2)**2)])

        
        ##########    Bisogna calcolare la propagazione degli errori per i parametri
    result = np.ones(9)
    
    result[0] = slope[0]
    result[1] = err_slope[0]
    result[2] = slope[1]
    result[3] = err_slope[1]
    result[4] = intercept[0]
    result[5] = err_int[0]
    result[6] = intercept[1]
    result[7] = err_int[1]
    result[8] = err
    
    return(result)
```

```python
fit = np.apply_along_axis(tracking, 1, Position_e_p.values[:1,],
                                [-0.21, -0.11, -0.01], np.array([[0,0], [0,0], [0,0]]), [0.2,0.2,0.2], precision_det=0.00005)

col = ['Sl_x', 'Err_Sl_x', 'Sl_y', 'Err_Sl_y', 'Int_x', 'Err_Int_x', 'Int_y', 'Err_Int_y', 'Sum.square']
fit_pos = pd.DataFrame(fit, columns=col)
fit_pos
```
