Bisogna scegliere la posizione dei detector prima del target e dobbiamo ottenere le componenti x e y dei positroni al variare della loro distanza dal target (il fascio di positroni non è allineato ma divergente). La seconda cosa può essere ottenuta simulando un boost laterale (da capire come fare), mentre la prima è a scelta.

Attenzione alle unità di misura. I dati sono in GeV, ma per fare dei boost o delle EOM ho bisogno di Joule o Nm.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import poisson
from skspatial.objects import Line, Points
from skspatial.plotting import plot_3d
import scipy.odr as spo


df=pd.read_csv("Documento 1 senza nome", sep=" ", index_col=False)
df
```

```python
mu_mass = 105.65/(10**3)
c = 299792458

NEvents=df.shape[0]

X=df["x_scattering"]
Y=df["y_scattering"]
Z=df["z_scattering"]

px_e=df["px_positron"]
py_e=df["py_positron"]
pz_e=df["pz_positron"]

Px_mu1=df["px_mu1"]
Py_mu1=df["py_mu1"]
Pz_mu1=df["pz_mu1"]

Px_mu2=df["px_mu2"]
Py_mu2=df["py_mu2"]
Pz_mu2=df["pz_mu2"]

Position_e_p = pd.DataFrame()
Position_mu_n = pd.DataFrame()
Position_mu_p = pd.DataFrame()
Position_mu_n_after_mag = pd.DataFrame()
Position_mu_p_after_mag = pd.DataFrame()
```

```python
#Z pos of the detectors before the target
Pos1= -0.21
Pos2= -0.11
Pos3= -0.01
```

```python
Z_Det_beforeTarg=[Pos1,Pos2,Pos3]

for i in Z_Det_beforeTarg:
    t=(i-Z)/(pz_e)   #array of "t" parameter, one for each event
    x=X+(px_e)*t    #element-wise operations
    y=Y+(py_e/mu_mass*c)*t
    
    d = np.sqrt(x**2+y**2)
    d_M = np.sort(d)[-1].mean()
    
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(18,8))
    fig.suptitle('Position distribution at Z={} meters before target'.format(-1*i),fontsize='x-large')
    
    #plot the histogram with the particle distribution
    h2d = ax1.hist2d(x,y,bins=(100,100))
    ax1.set_xlabel("Position along x (m)",fontsize='large')
    ax1.set_ylabel("Position along y (m)",fontsize='large')
#     ax1.set_xlim(-1.3*d_M, 1.3*d_M)
#     ax1.set_ylim(-1.3*d_M, 1.3*d_M)
    fig.colorbar(h2d[3], ax=ax1)
    
    #distribution with the detector shape
    ax2.scatter(x,y,color='red', marker='x')
    r=1.1 * d_M         #radius
    print(r)
#     ax2.scatter(xs,np.sqrt(r**2-xs**2),color='black',marker='.')
#     ax2.scatter(xs,-np.sqrt(r**2-xs**2),color='black',marker='.')
    ax2.grid(visible=True,which='both', linewidth=1)
    ax2.vlines(-r, -r, r)
    ax2.vlines(r, -r, r)
    ax2.hlines(-r, -r, r)
    ax2.hlines(r, -r, r)
    ax2.set_xlabel("Position along x (m)",fontsize='large')
    ax2.set_ylabel("Position along y (m)",fontsize='large')
    plt.show()
    Position_e_p['e+_X_{}'.format(-1*i)] = x
    Position_e_p['e+_Y_{}'.format(-1*i)] = y
    Position_e_p['e+_Z_{}'.format(-1*i)] = i*np.ones(NEvents)
    
Position_e_p.head(40)
```

```python
def tracking(name_part, pos_detectors, radious_detector, data):    # dovro considerare anche la diversa precisione dei detector
    
    N = 10
    
    result = np.zeros((N, 4))          # devo mettere 8 al posto di 4 se ho anche gli errori
    
    for i in range(N):
        k = poisson.rvs(mu=0, size=3)
        noise = np.vstack((np.random.uniform(-radious_detector[0], radious_detector[0], size=(k[0], 2)),
                          np.random.uniform(-radious_detector[1], radious_detector[1], size=(k[1], 2)),
                          np.random.uniform(-radious_detector[2], radious_detector[2], size=(k[2], 2))))
        
        hit, xed, yed, _ = plt.hist2d(noise[:, 0], noise[:, 1],
                                      bins=[np.arange(-radious_detector.max()-0.000025, radious_detector.max()+0.000025, 0.00005),
                                            np.arange(-radious_detector.max()-0.000025, radious_detector.max()+0.000025, 0.00005)])
        plt.close()

        center_bins_x = 0.5*(xed[:-1]+xed[1:])
        center_bins_y = 0.5*(yed[:-1]+yed[1:])

        mask = (hit.T[::-1,:]>0)
        X, Y = np.meshgrid(center_bins_x, center_bins_y[::-1])
        Z = np.array([X, Y])

        noise[:, 0] = np.round(Z[0][mask], 6)
        noise[:, 1] = np.round(Z[1][mask], 6)

        det_noise_pos = np.array(np.vstack((Z_Det_beforeTarg[0]*np.ones((k[0],1)),
                          Z_Det_beforeTarg[1]*np.ones((k[1],1)),
                          Z_Det_beforeTarg[2]*np.ones((k[2],1)))))

        detector_noise = np.array(np.hstack((noise, det_noise_pos)))

        for j in [0, 3, 6]:
            hit, xed, yed, _ = plt.hist2d(np.array([data.values[i,j]]), np.array([data.values[i,j+1]]),
                                          bins=[np.arange(-radious_detector.max()-0.000025, radious_detector.max()+0.000025, 0.00005),
                                            np.arange(-radious_detector.max()-0.000025, radious_detector.max()+0.000025, 0.00005)])
            plt.close()

            center_bins_x = 0.5*(xed[:-1]+xed[1:])
            center_bins_y = 0.5*(yed[:-1]+yed[1:])

            mask = (hit.T[::-1,:]>0)
            X, Y = np.meshgrid(center_bins_x, center_bins_y[::-1])
            Z = np.array([X, Y])
            
            data.values[i,j] = np.round(Z[0][mask], 6)
            data.values[i,j+1] = np.round(Z[1][mask], 6)
            
        hit_detector = np.array(data.values[i,:])

        signal_detector = np.vstack((detector_noise, hit_detector.reshape(3,3)))
        print(signal_detector, '\n')
        
        det_1 = signal_detector[signal_detector[:,2]==pos_detectors[0]]
        det_2 = signal_detector[signal_detector[:,2]==pos_detectors[1]]
        det_3 = signal_detector[signal_detector[:,2]==pos_detectors[2]]
        
        def f(B, x):
#         '''Linear function y = m*x + b'''
            return B[0]*x + B[1]
    
        sx = 0.00005
        sy = 0.00005
        sz = 0.00005
        
        err = 10**10
        slope = np.ones(3)
        intercept = np.ones(3)
        
        for n1 in range(det_1.shape[0]):
            for n2 in range(det_2.shape[0]):
                for n3 in range(det_3.shape[0]):
                    points = np.vstack((det_1[n1, :], det_2[n2, :], det_3[n3, :]))
                    x = points[:, 0]
                    y = points[:, 1]
                    z = points[:, 2]
                    
                    linear = spo.Model(f)

                    mydata_x = spo.RealData(x, y, sx=sx, sy=sy)
                    myodr_x = spo.ODR(mydata_x, linear, beta0=[1., 2.])
                    myoutput_x = myodr_x.run()

                    mydata_z = spo.RealData(z, y, sx=sz, sy=sy)
                    myodr_z = spo.ODR(mydata_z, linear, beta0=[1., 2.])
                    myoutput_z = myodr_z.run()
                    
                    err_sqrt = np.sqrt(myoutput_z.sum_square + myoutput_x.sum_square) 
                    
                    if err_sqrt<err:
                        err = err_sqrt
                        slope = np.array([myoutput_z.beta[0], myoutput_z.beta[0]/myoutput_x.beta[0], 1])
                        intercept = np.array([(myoutput_z.beta[1]-myoutput_x.beta[1])/myoutput_x.beta[0],
                                              myoutput_z.beta[1],
                                              0])
        
#         fig = plt.figure()
#         ax = fig.add_subplot(projection='3d')
#         ax.scatter(signal_detector[:,0], signal_detector[:,1], signal_detector[:,2], c='r')
#         t = np.linspace(pos_detectors.min(), pos_detectors.max(), 1000)
#         ax.plot(slope[0]*t+intercept[0], slope[1]*t+intercept[1], slope[2]*t+intercept[2])
#         ax.set_xlabel('X')
#         ax.set_ylabel('Y')
#         ax.set_zlabel('Z')
#         ax.view_init(0,0)
        
        
        ###########    Bisogna calcolare la propagazione degli errori per i parametri
        
        result[i, 0] = slope[0]
        result[i, 1] = slope[1]
        result[i, 2] = intercept[0]
        result[i, 3] = intercept[1]
        
    return(result)
        
        
        
        
        
matrice = tracking('Positroni', np.array([Pos1,Pos2,Pos3]), np.array([0.0110, 0.0110, 0.0110]), Position_e_p)
matrice
```

Adesso tocca ai muoni prima del magnete. Siccome abbiamo bustato i positroni, dobbiamo bustare anche i muoni per conservazione del quadrimomento. Inoltre dobbiamo valutare la distanza relativa tra i due muoni prodotti nello stesso evento: vogliamo che siano adeguatamente separati dai detector anche prima del magnete.

Ed ora è venuto il momento dei muoni dopo il campo magnetico. Chiaramente prima è necessario studiare l'effetto del campo magneto sui muoni, i quali entrano con un momento definito in un punto definito nello spazio. (consideriamo il punto x,y nel piano trasversale all'esperimento, mentre Z è fissato dalla posizione del magnete).

```python
B=1.7 #Tesla
L=3 #meters  
Z_mag = 8.03

#########################
#### Positive muons #####
#########################

NEvents=Px_mu2.shape[0]
#initial momenta and circ radiuses
P_pos=np.sqrt(Px_mu2**2+Pz_mu2**2)
R_pos=P_pos/(B*0.3)                 #meters
Circ_center=np.zeros((NEvents,2))

#initial positions
Z_init_pos=np.ones(NEvents)*Z_mag
t=(Z_init_pos-Z)/Pz_mu2

X_init_pos=X+Px_mu2*t
Y_init_pos=Y+Py_mu2*t


#Circ center
alpha=np.arccos(Pz_mu2/P_pos)
beta=np.pi/2-alpha

Circ_center[:,0][Px_mu2>=0]=(X_init_pos+R_pos*np.sin(beta))[Px_mu2>=0]
Circ_center[:,1][Px_mu2>=0]=(Z_init_pos-R_pos*np.cos(beta))[Px_mu2>=0]
   
Circ_center[:,0][Px_mu2<0]=(X_init_pos+R_pos*np.sin(beta))[Px_mu2<0]
Circ_center[:,1][Px_mu2<0]=(Z_init_pos+R_pos*np.cos(beta))[Px_mu2<0]

#positions after the magnet
Pos_mupos=np.zeros((NEvents,3))

Pos_mupos[:,0]= Circ_center[:,0]-np.sqrt(R_pos**2-((Z_init_pos+L)-Circ_center[:,1])**2)  # x
Pos_mupos[:,2]= Z_init_pos+L    # z

string=np.sqrt((Pos_mupos[:,0]-X_init_pos)**2+(Pos_mupos[:,2]-Z_init_pos)**2)
theta=2*np.arcsin(string/(2*R_pos))
theta=2*np.arcsin(3/(2*R_pos))


k=R_pos*theta/P_pos

Pos_mupos[:,1]=Py_mu2*k+Y_init_pos    # y  
#momenta
Momenta_mupos=np.zeros((NEvents,3))


Momenta_mupos[:,0]=np.cos(theta/2)*Px_mu2+np.sin(theta/2)*Pz_mu2
Momenta_mupos[:,1]=py_mu2
Momenta_mupos[:,2]=np.cos(theta/2)*Pz_mu2-np.sin(theta/2)*Px_mu2

print(Momenta_mupos)
```

```python
##solve same equations as before to get where they hit final detectors after magnet
Z_Det_afterMagn=[12.43, 13.43, 14.43] 
#parameters of detector shape, used  to check if detector's sizes are accurate
x_center=[0.056,0.103,0.15]
a=[0.047,0.064,0.082] #x semiaxis
b=[0.037,0.04,0.043] #y semiaxis

for i,xi,ai,bi in zip(Z_Det_afterMagn,x_center,a,b):
    #negative muons
    tn=(i-Pos_muneg[:,2])/Momenta_muneg[:,1]   #array of"t" parameter,one for each event
    xn=Pos_muneg[:,0]+Momenta_muneg[:,0]*tn    #element-wise operations
    yn=Pos_muneg[:,1]+Py_mu1*tn
    
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(14,10))
    fig.suptitle('Muons positions along the trasversal plane at Z={} meters from target'.format(i), fontsize=14)
    ax1.scatter(xn,yn,color='blue',marker='x')
    ax1.set_xlabel("Position along x (m)")
    ax1.set_ylabel("Position along y (m)")
    #draw the detector shape
    xs=np.linspace(xi-ai,xi+ai,1000)
    ax1.scatter(xs,bi*np.sqrt(1.001-(xs-xi)**2/ai**2),color='black',marker='.')
    ax1.scatter(xs,-bi*np.sqrt(1.001-(xs-xi)**2/ai**2),color='black',marker='.')
    ax1.grid(b=True,which='both', linewidth=1)
    h2d = ax2.hist2d(xn,yn,bins=(150,150),cmap='inferno')
    fig.colorbar(h2d[3], ax=ax2)
    plt.grid()
    ax2.set_xlabel("Position along x (m)")
    ax2.set_ylabel("Position along y (m)")
    plt.show()

    #positive muons
    tp=(i-Pos_mupos[:,2])/Momenta_mupos[:,1] 
    xp=Pos_mupos[:,0]+Momenta_mupos[:,0]*tp
    yp=Pos_mupos[:,1]+Py_mu2*tp
    
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(14,10))
    fig.suptitle('Antimuons positions along the trasversal plane at Z={} meters from target'.format(i),fontsize=14)
    ax1.scatter(xp,yp,color='red',marker='+')
    ax1.grid(b=True,which='both', linewidth=1)
    ax1.set_xlabel("Position along x (m)")
    ax1.set_ylabel("Position along y (m)")
    #draw the detector shape
    xs=np.linspace(-xi-ai,-xi+ai,1000)
    ax1.scatter(xs,bi*np.sqrt(1.001-(xs+xi)**2/ai**2),color='black',marker='.')
    ax1.scatter(xs,-bi*np.sqrt(1.001-(xs+xi)**2/ai**2),color='black',marker='.')
    h2d = ax2.hist2d(xp,yp,bins=(150,150),cmap='inferno')
    fig.colorbar(h2d[3], ax=ax2)
    plt.grid()
    ax2.set_xlabel("Position along x (m)")
    ax2.set_ylabel("Position along y (m)")
    plt.show()
```

Definizione della funzione che si occupa del tracciare l'esperimento. Questa restituisce solo gli hits.

```python
def hits(total_datf,numberf):
    
    #splitting input matrix for convenience
    (Xf, Yf, Zf, Px_ef, Py_ef, Pz_ef, Px_mu1f, Py_mu1f, Pz_mu1f, Px_mu2f, Py_mu2f, Pz_mu2f,
            Pos_muneg0f, Pos_muneg1f, Pos_muneg2f, Momenta_muneg0f, Momenta_muneg1f,
            Pos_mupos0f, Pos_mupos1f, Pos_mupos2f, Momenta_mupos0f, Momenta_mupos1f) = (total_datf)
    
    #number of external signals due to noise
    numberf=numberf
    
    #posizione del detector
    Z_det = np.array([-0.21,-0.11,-0.01,
                      5.03,6.03,7.03,
                      9.43,10.43,11.43])
    
    #punto di riferimento del detector
    det_XY = np.array([[-0.011,-0.011], [-0.011,-0.011], [-0.011,-0.011],
                       [-0.025,-0.025], [-0.028,-0.028], [-0.03,-0.03],
                       [0.009,-0.037],  [0.039,-0.04],   [0.068,-0.043]])
    
    #dimensioni laterali del detector
    det_sides = np.array([[0.022, 0.022], [0.022, 0.022], [0.022, 0.022],
                          [0.05, 0.05], [0.056, 0.056], [0.06, 0.06],
                          [0.094, 0.074], [0.128, 0.08],[0.164, 0.086]])
    
    #lists used for saving coordinates of active pixels' centers (i.e. both noise and signal)
    #in every list there will be three elements, one for each detector of each bunch
    #(before target, between target and magnetic field, after magnetic field "right" and "left")
    #every element will be a list of coordinates, so 2-elements lists
    hit_raw_before = []
    hit_raw_between = []
    hit_raw_aftern = []
    hit_raw_afterp = []
    
    """
    Simulation before the target - positrons
    """
    
    for k in range(0,3):
        #EOM
        i=Z_det[k]
        t=(i-Zf)/Pz_ef   
        x_sim=Xf+Px_ef*t
        y_sim=Yf+Py_ef*t
        
        #assignment of new names for convenience
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[k][0],det_XY[k][1],det_sides[k][0],det_sides[k][1],50e-6,x_sim,y_sim)
        
        
        #number is a matrix with poisson casual numbers that represent the number n of active pixels
        #in a single detector due to noise
        #so we extract n couple of integer numbers that are the coordinates of the noisy pixels
        #expressed in pixels-unit
        
        temp1=np.random.randint(1,np.int(sideX/pitch),numberf[k])
        temp2=np.random.randint(1,np.int(sideY/pitch),numberf[k])
        
        #Ora vogliamo esprimere le coordinate dei pixel rumorosi in metri
        pixels = [[0.1, 0.1] for i in range(len(temp1))] #dummy pixels list of the right length
        
        for i in range(0,len(temp1)):
            #saving coordinates of noisy pixels
            pixels[i] = [X0+temp1[i]*pitch-0.5*pitch,Y0+temp2[i]*pitch-0.5*pitch]
        
        #Ora viene l'esperimento vero e proprio. Qui facciamo passare N positroni e verifichiamo se passano
        #nel detector o meno.
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            #checking for particle being within detector
            pixels.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        #elenco pixel eccitati per ogni detector (ogni riga dell'array)
        hit_raw_before.append(pixels)
        
        
    """
    Simulation between the target and the magnet
    """
    
    for k in range(3,6):
        i=Z_det[k]
        
        tn=(i-Zf)/Pz_mu1f   
        xn=Xf+Px_mu1f*tn    
        yn=Yf+Py_mu1f*tn
        
        tp=(i-Zf)/Pz_mu2f
        xp=Xf+Px_mu2f*tp
        yp=Yf+Py_mu2f*tp
        
        ### negative muons ###
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[k][0],det_XY[k][1],det_sides[k][0],det_sides[k][1],50e-6,xn,yn)
        
        temp1=np.random.randint(1,np.int(sideY/pitch),numberf[k])
        temp2=np.random.randint(1,np.int(sideX/pitch),numberf[k])
        pixels1 = [[0.1, 0.1] for i in range(len(temp1))]
        
        for i in range(0,len(temp1)):
            pixels1[np.intp(i)] = [X0+temp2[i]*pitch-0.5*pitch,Y0+temp1[np.intp(i)]*pitch-0.5*pitch]
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels1.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        ### positive muons ###
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[k][0],det_XY[k][1],det_sides[k][0],det_sides[k][1],50e-6,xp,yp)
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels1.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        #elenco pixel eccitati per ogni detector (ogni riga dell'array)
        hit_raw_between.append(pixels1)
          
      
    """
    Simulation between the target and the magnet
    """
    
    #mu-
    for k in range(6,9):
        i=Z_det[k]
        tn=(i-Pos_muneg2f)/Momenta_muneg1f
        xn=Pos_muneg0f+Momenta_muneg0f*tn
        yn=Pos_muneg1f+Py_mu1f*tn
        
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(det_XY[np.intp(k)][0],det_XY[np.intp(k)][1],det_sides[np.intp(k)][0],det_sides[np.intp(k)][1],50e-6,xn,yn)
        temp1=np.random.randint(1,np.int(sideX/pitch),numberf[np.intp(k)])
        temp2=np.random.randint(1,np.int(sideY/pitch),numberf[np.intp(k)])
        pixels2 = [[0.1, 0.1] for i in range(len(temp1))]
        
        for i in range(0,len(temp1)):
            pixels2[np.intp(i)] = [X0+temp1[np.intp(i)]*pitch-0.5*pitch,Y0+temp2[np.intp(i)]*pitch-0.5*pitch]
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels2.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        hit_raw_aftern.append(pixels2)
    
    #mu+
    for k in range(6,9):
        i=Z_det[k]
        tp=(i-Pos_mupos2f)/Momenta_mupos1f
        xp=Pos_mupos0f+Momenta_mupos0f*tp
        yp=Pos_mupos1f+Py_mu2f*tp
        
        (X0, Y0, sideX, sideY, pitch, xpart, ypart)=(-det_XY[np.intp(k)][0]-det_sides[np.intp(k)][0],det_XY[np.intp(k)][1],det_sides[np.intp(k)][0],det_sides[np.intp(k)][1],50e-6,xp,yp)
        temp1=np.random.randint(1,np.int(sideY/pitch),numberf[np.intp(k)])
        temp2=np.random.randint(1,np.int(sideX/pitch),numberf[np.intp(k)])
        pixels3 = [[0.1, 0.1] for i in range(len(temp1))]
        
        for i in range(0,len(temp1)):
            pixels3[np.intp(i)] = [X0+temp2[np.intp(i)]*pitch-0.5*pitch,Y0+temp1[np.intp(i)]*pitch-0.5*pitch]
        
        if xpart>X0 and ypart>Y0 and xpart<(X0+sideX) and ypart<(Y0+sideY) :
            pixels3.append([X0+np.int(np.abs(xpart-X0)/pitch)*pitch+0.5*pitch,
                           Y0+np.int(np.abs(ypart-Y0)/pitch)*pitch+0.5*pitch])
        else:
            print("particle out of detector")
        
        hit_raw_afterp.append(pixels3)
        
    return hit_raw_before, hit_raw_between, hit_raw_aftern, hit_raw_afterp
```
