### Esercizio 1

```python
import numpy as np
import math
dimension = 200
distance = np.zeros((dimension,1))
for x in range(1,dimension+1):
    steps = np.random.randint(2, size=(1000,20*x))
    steps[steps==0] = -1
#     print(steps)
#     print(steps.shape)
    tot_dist = np.sum(steps, axis=1)
    tot_dist = tot_dist**2
    tot_dist = np.sqrt(tot_dist)
#     print(tot_dist.shape)
#     print(tot_dist)
    distance[x-1] = tot_dist.mean()
distance
```

### Esercizio 2

```python
import numpy as np
import math
dimension = 200
distance = np.zeros((dimension,1))
for x in range(1,dimension+1):
    steps = np.random.randint(2, size=(1000,20*x))
    steps[steps==0] = -1
    tot_dist = np.sum(steps, axis=1)
    distance[x-1] = int(tot_dist.mean())
print(distance)
```
