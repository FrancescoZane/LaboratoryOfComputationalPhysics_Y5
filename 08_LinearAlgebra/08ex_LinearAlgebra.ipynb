    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$

3 \. **Looking at an oscillating spring** (optional)

Imagine you have $n$ cameras looking at a spring oscillating along the $x$ axis. Each  camera record the motion of the spring looking at it along a given direction defined by the pair $(\theta_i, \phi_i)$, the angles in spherical coordinates. 

Start from the simulation of the records (say ${\cal O}(1000)$) of the spring's motion along the x axis, assuming a little random noise affects the measurements along the $y$. Rotate such dataset to emulate the records of each camera.

Perform a Principal Component Analysis on the thus obtained dataset, aiming at finding the only one coordinate that really matters.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
import pandas as pd

N = 2000
t = np.linspace(0,1,N)
x = 2 * np.sin(2*t*np.pi)
y = 0.2*np.random.uniform(-1, 1, N)
z = np.zeros(N)
d = np.array([x, y, z])

pos = np.array([x, y, z])

def rot_x(t):
    matrix = np.array([[1, 0, 0],
                      [0, np.cos(t), -np.sin(t)],
                      [0, np.sin(t), np.cos(t)]])
    return matrix
    
def rot_y(p):
    matrix = np.array([[np.cos(p), 0, np.sin(p)],
                      [0, 1, 0],
                      [-np.sin(p), 0, np.cos(p)]])
    return matrix
    
def camera(t, p):
    print('\n','La camera definita dagli angoli:   [ theta , phi ] : [', np.round(t,2), ',', np.round(p,2), ']', '\n')
    n_d = np.dot(np.dot(rot_x(-t), rot_y(-p)), d)
    l, V = la.eig(np.cov(n_d))
    ris = {'Eigenvector': [np.round(V[:,i], 2) for i in range(3)],
           'Eigenvalue': [np.round(np.real(l[i]), 3) for i in range(3)],
           'Var info in(%)': [np.round(np.real(l[i]/np.sum(l))*100, 3) for i in range(3)]          
          }
    dati = pd.DataFrame(data=ris, index=['Vect 1', 'Vect 2', 'Vect 3'])
    print(dati, '\n')
    
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(projection='3d')
    ax.scatter(n_d[0,:], n_d[1,:], n_d[2,:], alpha=0.02)
    ax.set_xlabel('Camera x axis')
    ax.set_ylabel('Camera y axis')
    ax.set_zlabel('Camera z axis')
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_zlim(-2, 2)
    ax.scatter(0, 0, 0, color='red')
    for i in range(3):
        ax.plot([0,np.real(l[i])*V[0,i]], [0,np.real(l[i])*V[1,i]], [0,np.real(l[i])*V[2,i]], color='red')
    plt.show()
    
    
    
    
camera(0,0)
camera(0, np.pi/2)
```
